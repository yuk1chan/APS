```julia
function psi(r,a)
    return exp(-r^2/a)
end
```

```julia
function psi2(r,a)
    return exp(-2(r^2)/a)
end
```

```julia
function V(r,α)
    return exp(-α*r)/r
end
```

```julia
function psiHpsir2(r,a,α)
    return  (-1/2)*( (-4(r^2)/a)psi2(r,a) + ((2r^4)/a^2)psi2(r,a) ) - V(r,α)*psi2(r,a)*r*r
end
```

```julia
function point(δ)
    return δ*(2*rand()-1)
end
```

```julia
function calcψHψr2(N,a)
    α = 5
    β = -1/2
    S = α*β
    count = 0
    for i in 1:N
        
        r = α*rand()
        y = β*rand()
               
        if y >= psiHpsir2(r,a,1.0)
            count += 1
        end
    end
    
    S = S*count/N
    
    return S
    
end
```

```julia
function calcψψr2(N,a)
    α = 10
    β = 1
    S = α*β
    count = 0
    for i in 1:N
        
        r = α*rand()
        y = β*rand()
        
        if y <= psi2(r,a)*r*r
            count += 1
        end
    end
    
    S = S*count/N
end
```

```julia
function Monte(N,a)

    
    psiHpsi = calcψHψr2(N,a)
    psipsi = calcψψr2(N,a)
    
    E = psiHpsi/psipsi
    
    println("psiHpsi = $psiHpsi")
    println("psipsi = $psipsi")
    println("E = $E")
    println()
end
```

```julia
function main()
    
    N = 10000000
    α = 1/10
    for i in 500:1:550
        @show a = α*i
        Monte(N,a)
    end
    
end
main()
```
