```julia
function psi(r,a)
    return exp(-r/a)
end

function psi2(r,a)
    return exp(-2r/a)
end
```

```julia
diff(psi(r,a),r,2)
```

```julia
function V(r)
    return -1/r
end
```

```julia
function psiHpsir2(r,a)
    return  (-1/2)*( (-2r/a)*psi2(r,a) + ((r^2)/a^2)*psi2(r,a) ) + V(r)r*r*psi2(r,a)
end
```

```julia
function point(δ)
    return δ*(2*rand()-1)
end
```

```julia
function calcψHψr2(N,a)
    α = 5
    β = -1/2
    S = α*β
    count = 0
    for i in 1:N
        
        r = α*rand()
        y = β*rand()
               
        if y >= psiHpsir2(r,a)
            count += 1
        end
    end
    
    S = S*count/N
    
    return S
    
end
```

```julia
function calcψψr2(N,a)
    α = 10
    β = 1
    S = α*β
    count = 0
    for i in 1:N
        
        r = α*rand()
        y = β*rand()
        
        if y <= psi2(r,a)*r*r
            count += 1
        end
    end
    
    S = S*count/N
end
```

```julia
function Monte(N,a)

    
    psiHpsi = calcψHψr2(N,a)
    psipsi = calcψψr2(N,a)
    
    E = psiHpsi/psipsi
    
    println("psiHpsi = $psiHpsi")
    println("psipsi = $psipsi")
    println("E = $E")
    println()
end
```

```julia
function main()
    
    N = 10000000
    α = 1/10
    for i in 0:1:20
        @show a = α*i
        Monte(N,a)
    end
    
end
main()
```
